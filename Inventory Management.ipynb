## Product Details
1. Product ID
2. Product Name
3. Price of Product
4. Quantity

```python
#Noting the time of transaction
import time

#Reading the inventory
fd=open('Inventory.txt','r')
products=fd.read().split('\n')
fd.close()

#Taking user I/P
u_name       =input("Enter your username: ")
u_email      =input("Enter your Email-ID: ")
u_phone      =input("Enter your phone no: ")
prod_id      =input("Enter product ID: ")
prod_quantity=input("Enter quantity: ")

updated_product_lst=[]

#Going through each product detail
for product in products:
    
    prod_details=product.split(',')
    
    # Checking if product exists or not
    if(prod_details[0]==prod_id):
        
        # If we're having enough quantity
        if(int(prod_quantity) <= int(prod_details[3])):
            
            print('------------------------------------------------------------------------------------------------------------')
            print('Name of the Product           : ',prod_details[1])
            print('Price of the Product          : ',prod_details[2])
            print('Quantity of the product       : ',prod_quantity)
            print('------------------------------------------------------------------------------------------------------------')
            print('Billing Amount                : ',int(prod_quantity)*int(prod_details[2]))
            print('------------------------------------------------------------------------------------------------------------')
            print('------------------------------------------------------------------------------------------------------------')

            #to update the quantity in the inventory
            prod_details[3]=str(int(prod_details[3])-int(prod_quantity))
            
            #Generating Sales in Sales.txt
            fd=open('Sales.txt','a')
            sales_detail= u_name+"," + u_email + "," + u_phone + "," + prod_details[1] + "," + prod_id + "," + prod_quantity + "," + str(int(prod_quantity)*int(prod_details[2])) + "," + time.ctime() +"\n"
            fd.write(sales_detail)
            fd.close()
        
        else:
            
            # If we're not having enough quantity
            print("Sorry, We're not having enough quantity.")
            print("We're having only",prod_details[3],"quantity.")
            print("Would you like to purchase it?")
            
            ch=input("Press Y/N")
            
            # If you want to purchase with remaining quantity
            if(ch=="Y" or ch=="y"):
                print('------------------------------------------------------------------------------------------------------------')
                print('Name of the Product           : ',prod_details[1])
                print('Price of the Product          : ',prod_details[2])
                print('Quantity of the product       : ',prod_details[3])
                print('------------------------------------------------------------------------------------------------------------')
                print('Billing Amount                : ',int(prod_details[3])*int(prod_details[2]))
                print('------------------------------------------------------------------------------------------------------------')
                print('------------------------------------------------------------------------------------------------------------')

               
                
                #Generating Sales in Sales.txt
                fd=open('Sales.txt','a')
                sales_detail= u_name+"," + u_email + "," + u_phone + "," + prod_details[1] + "," + prod_id + "," + prod_details[3] + "," + str(int(prod_details[3])*int(prod_details[2])) + "," + time.ctime() +"\n"
                fd.write(sales_detail)
                fd.close()
                
                #to update the quantity in the inventory
                prod_details[3] = '0'
            else:
                
                print("Thanks")
                
    # Updating my Inventory List            
    updated_product_lst.append(prod_details)
    

#from here onwards we are doing to show the required o/p in the file only
lst = []

# Updating my Inventory String
for i in updated_product_lst:
    prod = i[0] +","+  i[1] +","+ i[2] +","+ i[3] + '\n'
    lst.append(prod)

lst[-1] = lst[-1][:-1] #this is done to remove the extra \n(enter) at the end of every updated list
  
# Updating Inventory File    
fd = open('Inventory.txt','w')

for i in lst:
    fd.write(i)

fd.close()

print("-------------------")
print("Inventory Updated")
```
